# Mission 1

Define a list, then display only the elements (and their position) that have even positions (0, 2, 4, 6...).

With `test_list = ["a", "b", "c", "d", "e"]`, the script should display:
```
a at position 0
c at position 2
e at position 4
```

````
test_list = ["a", "b", "c", "d", "e"]

for i, val in enumerate(test_list):
  if i %2 == 0 :
    print(f"{val} at position {i}")

````

# Mission 6

Ask the user for a string, convert it to a list of characters with the `list` function, then ask for a single character.

Remove the first occurrence of the character from the list, then display how many occurrences of the character remain in the list.

For instance if you enter `"Excellent work"` then the letter `"e"`, it will display:
```
There are still 2 copies of e in the list
```

````
element= str(input("Ecrire ce que tu veux ! : "))
lettre= str(input("Ecrire un caractére que l'on souhaite ! : "))
element= list(element)
print(f"il y a {element.count(lettre)} fois la lettre {lettre} dans la liste")
````

````
phrase= list(input("Ecrire ce que tu veux ! : ").lower())
lettre= str(input("Ecrire un caractére que l'on souhaite ! : ").lower())
nb_phrase =len(phrase)

for l in range(nb_phrase) :
  if phrase[l] == lettre :
    del phrase[l]
    break
count=0
nb_phrase =len(phrase)
for x in range(nb_phrase):
  if phrase[x] == lettre :
    count+=1
print(f"il y a {count} fois la lettre {lettre} dans la liste")
````

# Mission 7

Define three lists containing no duplicates. Then, indicate (however you like) the two that have the most elements in common (or all three if there's a tie).

For instance, with:
```
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]
```
It may display:
```
Lists B and C have the most common elements!
```

````
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]

comp_1=len(set(list_a) & set(list_b))
comp_2=len(set(list_b) & set(list_c))
comp_3=len(set(list_a) & set(list_c))

if comp_1>comp_2 and comp_1>comp_3:
  print(f"la liste A et la liste B ont le plus d'elements en commun")
if comp_2>comp_1 and comp_2>comp_3:
  print(f"la liste B et la liste C ont le plus d'elements en commun")
if comp_3>comp_2 and comp_3>comp_1:
  print(f"la liste A et la liste C ont le plus d'elements en commun")
````
