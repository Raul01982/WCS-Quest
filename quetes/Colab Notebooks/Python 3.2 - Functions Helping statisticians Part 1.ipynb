```
# Mission 4: Create a function that takes a string as input, and return its vowels only, preserving case.
# Example: "Antidisestablishmentarianism" -> "Aiieaieaiai"
# Example: "TO BE OR NOT TO BE" -> "OEOOOE"

def garde_v (var_1):

  l_phrase=list(var_1)
  nb_phrase=len(var_1)
  voyelle = ["a","e","y","i","o","u"]
  count=0

  for x in range(nb_phrase) :
    for l in range((x+1)- count) :
      if l_phrase[l].lower() not in voyelle:
        del l_phrase[l]
        count+=1
        break

  return "".join(l_phrase)

phrase= input("Ecrire ce que tu veux ! : ")

garde_v(phrase)
```

```
def garde_v (var_1):

  l_phrase=list(var_1)
  nb_phrase=len(var_1)
  voyelle = ["a","e","y","i","o","u"]
  new_phrase=[]

  for l in range(nb_phrase):
    if l_phrase[l].lower() in voyelle:
       new_phrase.append(l_phrase[l])

  return "".join(new_phrase)

phrase= input("Ecrire ce que tu veux ! : ")

garde_v(phrase)
```
