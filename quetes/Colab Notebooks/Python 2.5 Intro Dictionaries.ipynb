For each cell, execute the code and answer the question before moving on to the next.

```python
people = {"Gontran": 23,
          "Jason": 42,
          "Ayoub": 27}


for i in people:
  print(i)
```

1. What does `i` correspond to?
To answer, you must use the terms "keys" and "values".

Answer: "i" definie la premiere partie de l'element soit les clés (keys) du dictionnaire "people"

```python
for i in people.keys():
  print(i)
```

```python
for i in people.values():
  print(i)
```

2. What does `i` correspond to in the second last cell?  
What does `i` in the last cell correspond to?  
Why do you ask?

Answer: dans les 2 boucles on a precisé si l'on rechercher la clés(keys) ou la valeur(values) donc "i" fait referance à soit la cles soit la valuer

```python
for i in people.items():
  print(i)
```

3. What is the type of object returned here called?

Answer: .items est là pour preciser que l'on recherche l'element entier donc la boucle fait apparaitre des tuples !

```python
for i, age in people.items():
  print(age)
```

# Mission 2



```python
people = {"Gontran": 23,
          "Jason": 42,
          "Ayoub": "27",
          "Léo": [10,20,30],
          "Rosario": {"Juan": 3,
                      "Camelia": 6}}


for i in people.values():
  print(type(i))
```

1. Does the above code allow me to confirm that I can have values of different types within a dictionary?

Answer: oui puisque le code nous affiche le type de chaque valeur du dictionnaire et on peut voir qu'il y des int, str, list, dict


2. Can I have keys of different types within a dictionary?


Answer: les cles sont une liste de caractere dans ce cas c'est un str mais cela peut aussi etre d'un autre type

```python
people["Ayoub"] = int(people["Ayoub"])

for i in people.values():
  print(type(i))
```

3. What did the above code allow me to do?

Answer: ce code permet de modifier le type de la valeur correspondant à la cles "Ayoub"

# Mission 3

```python
groups = {"groupe_1": ["Lam", "Ghizlaine", "Khaled", "Florian"],
          "groupe_2": ["Lucile", "Mbaye", "Cécile", "Rohan"],
          "groupe_3": ["Agathe", "Charlotte", "Charles", "Maxime"],
          "groupe_4": ["Gaelle", "Linh", "Meral"]}


for i in groups.values():
  print(i)
```

1. What is the type of each value?  
How do I get into each list?

Answer: les values sont des listes et je peux rentrer dans chaque liste en utilisant la cles et l'indice de la liste

```python
for i in groups.values():
  for v in i:
    print(v)
```

2. How did I get into each list?  
Take the time to explain your reasoning for answering this question.

Answer: on est rentrée dans chaque liste en faisant une boucle dans une boucle. Dans la premiere boucle "i" va chercher les valeurs qui sont des listes et dans la 2eme boucle "v" va chercher chaque element des listes
