For each cell, execute the code and answer the question before moving on to the next.

```python
names = ["Fredy", "Nicolas", "Alexandre"]

for i in enumerate(names):
  print(i)
```

1. What type of object is returned as output?  
Hint: You can use `type(i)`.

Answer : La méthode enumerate() permet de convertir une séquence en un objet contenant les mêmes éléments avec leur index correspondant sous forme de tuple. Le type de (i) est un element (index + valeur)


```python
for i, val in enumerate(names):
  print(i)
```

```python
for i, val in enumerate(names):
  print(val)
```

2. What does `i` correspond to?  
What does `val` correspond to?  
Why is it so?

Answer : (i) est l'index et (val) est la valeur associée, parceque la fonction enumerate() crée un index qui est appeler par la premiere variable de notre boucle

```python
for val, i in enumerate(names):
  print(i)
```

```python
for val, i in enumerate(names):
  print(val)
```

3. This time, `i` returns first names and `val` returns numbers. Why is it so?

Answer : On a changé les denomination de nos variable dans la boucle, la premiere etant l'index et la seconde la valeur.

4. What does the built in function `enumerate` do?
In what cases can it be useful?

Answer : la fonction enumerate() crée un index pour chaque element de la liste, cela permet de pouvoir positioner facilement un element dans la liste et en suite de le modifier
