# Mission 4

Create 2 tuples of the same length that contain only integers.

Then create a script that compares the sum of their elements and displays the tuple that has the highet total.

```
my_tuple_int_1=(15,25,89,65,74,23,19)
my_tuple_int_2=(65,85,45,28,76,37,91)
count_1=0
count_2=0
for x in my_tuple_int_1:
  count_1+=x
print(f"le totale de my_tuple_int_1 est : {count_1}")
for x in my_tuple_int_2:
  count_2+=x
print(f"le totale de my_tuple_int_2 est : {count_2}")
if count_1>count_2:
  print("my_tuple_int_1 est superieur à my_tuple_int_2")
if count_2>count_1:
  print("my_tuple_int_2 est superieur à my_tuple_int_1")
```
