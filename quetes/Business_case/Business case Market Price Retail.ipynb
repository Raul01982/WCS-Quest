```python
import pandas as pd
import datetime as dt
```

```python
df_market = pd.read_csv("https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/Market_retail.csv")
```

```python
df_market
```

```python
df_market.info()
```

```python
df_market['Order Date'] = df_market['Order Date'].astype('datetime64[ns]')
```

```python
df_market[' Discount '] = df_market[' Discount '].apply(lambda x : x.replace('$',''))
df_market[' Discount '] = df_market[' Discount '].apply(lambda x : x.replace('-','0'))

df_market[' Discount '].value_counts()
```

```python
df_market[' Discount '] = df_market[' Discount '].astype(float)
```

```python
df_market[' Sales '] = df_market[' Sales '].apply(lambda x : x.replace('$','').replace(',','').replace('-','0'))
df_market[' Profit '] = df_market[' Profit '].apply(lambda x : x.replace('$','').replace(',','').replace(' - ','0'))


df_market[' Sales '] = df_market[' Sales '].astype(float)
df_market[' Profit '] = df_market[' Profit '].astype(float)
```

```python
df_market[' Profit '].value_counts()
```

```python
df_market[' Sales '].value_counts()
```

```python
df_market
```
